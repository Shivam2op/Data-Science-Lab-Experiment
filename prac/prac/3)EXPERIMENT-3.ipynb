# Calculate summary statistics grouped by the categorical variable

```python
summary_stats = data.groupby("age")["income"].agg(["mean", "median", "min", "max", "std"]).reset_index()

# Create a list of numeric values for each response to the categorical variable
category_values = summary_stats["age"].tolist()

```
