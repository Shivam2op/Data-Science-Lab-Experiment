```python
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
numeric_data = df[numeric_columns]

# Identify outliers
outliers = pd.DataFrame(columns=numeric_data.columns)

for column in numeric_data.columns:
    lower_threshold = numeric_data[column].quantile(0.25) - 1.5 * (numeric_data[column].quantile(0.75) - numeric_data[column].quantile(0.25))
    upper_threshold = numeric_data[column].quantile(0.75) + 1.5 * (numeric_data[column].quantile(0.75) - numeric_data[column].quantile(0.25))
    column_outliers = numeric_data[(numeric_data[column] < lower_threshold) | (numeric_data[column] > upper_threshold)]
    outliers = pd.concat([outliers, column_outliers])

print("Identified outliers:")
print(outliers)
```
